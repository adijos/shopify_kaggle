## Conclusions and Next Steps

Based on the above analysis of looking at the population of apps and their average ratings, I noted that there are two areas to investigate further with potential for a new e-commerce app: Places to Sell and Orders and Shipping. Because there is significantly less competition in the former category as compared to the latter, I conclude that the new e-commerce app should be in the category of Places to Sell. 

Because this is only a preliminary analysis, there are a number of next steps that have to be taken to refine this opportunity. First, we need to understand what difficulties caused the majority of merchants to rate the existing apps with a less than satisfactory rating. This could amount to a text analysis of the reviews themselves or even just reading and classifying reviews manually. Either way, the content of the reviews is critical to understanding the context behind each merchant's submitted rating.

Second, we need to understand the operational cost of setting up a successful Places to sell app. The most installed apps in this category are integrations with existing brand-name firms. A new app in this area might offer better data tracking across existing Places to sell integrations, an easier to use interface to set up Places to Sell integrations, a way to aggregate and manage orders from different points of sale, or something else entirely based on more customer/market research. 

## Simplifying Assumptions

In this section, I want to highlight a number of assumptions I used to simplify the analysis for the sake of time and/or simplicity.

- [Reviewers only make up about 5-10% of a total customer population on average](https://websitebuilder.org/blog/online-review-statistics/#:~:text=About%205%25%E2%80%9310%25%20of,shoppers%20specifically%20seek%20negative%20reviews.). This means that all the customer ratings and reviews would skew towards the opinions of only the most engaged users (merchants). Because this dataset did not include overall downloads, I was unable to generalize further than this set of users.
- I am not familiar with the operational cost differences of setting up a business behind the apps of each category. For example, it might turn out that Trust and security apps require significant legal knowledge, skills outside of my individual abilities and ones that I would probably have to hire externally for, raising the cost of proceeding with this area of opportunity.
- Users might rate/review apps that are multifunctional (cover multiple app categories) in a certain way because of one particular functionality and not the app as a whole. This is obfuscated in the single number rating. I chose not to dig into users' reactions to overlapping apps for time's sake.
